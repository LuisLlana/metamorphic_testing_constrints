```python
import pandas as pd
import re
from io import StringIO
import os
```

```python
def aggregate(filename: str) -> None:
    with open(filename) as f:
        contents = f.read()
        contents, n  = re.subn('mut-', '', contents)
        print(f'{n} cambios')
        contents, n  = re.subn(r'-\d;', ';', contents)
        print(f'{n} cambios')
    df = pd.read_csv(StringIO(contents), sep=';')
    df1 = df.groupby(['Mutant']).max()
    new_name = filename.replace('killed', 'operators_killed')
    df1.to_csv(new_name, sep=';')
    
```

```python
files = ['V8_killed_different_outputs.csv',
    'V8_killed_MR1.csv',
    'V8_killed_MR2.csv',
    'V8_killed_MR3.csv',
    'V8_killed_MR4.csv']
```

```python
for file in os.listdir():
    if file.startswith('V8_killed') and file.endswith('.csv'):
        aggregate(file)
```
